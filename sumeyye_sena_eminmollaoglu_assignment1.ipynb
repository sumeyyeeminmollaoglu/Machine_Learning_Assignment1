```python
import pandas as pd
import numpy as np
```

## DATA EXPLORATION & PREPARATION

### part a

```python
df= pd.read_csv('train.csv')     #creating dataframe for train datas
train_x_a = df.drop("SalePrice",1)    #SalePrice column is dropped
train_y = pd.DataFrame(df["SalePrice"])   #creating new dataframe for SalePrice column 
```

```python
train_x_a.head(20)
```

```python
train_y.head(20)
```

### part b

```python
nan_columns = train_x_a.columns[train_x_a.isnull().any()].tolist()   #finding column names which have missing values
median_val={}      #dictionary for nan columns and their median values to use later
for i in nan_columns:
    median = train_x_a[i].median()   #median of the column
    median_val[i]=median
    train_x_a[i].fillna(median, inplace=True)    #filling the missing value cells with median of the column
train_x_b=train_x_a 
print (median_val)
```

```python
print(nan_columns)
```

```python
train_x_b.head(20)
```

### part c

```python
train_x_b_num = train_x_b.select_dtypes(include=[np.number])     #finding column names which contain numerical values
categorical_columns=list(set(train_x_b.columns) - set(train_x_b_num))    #finding categorical column names by subtracting numerical column names from all column names
```

```python
print(categorical_columns)
```

### part d

```python
from sklearn.preprocessing import OneHotEncoder
train_x_b_cat=train_x_b[categorical_columns]     
cat_encoder = OneHotEncoder(categories="auto")
train_x_b_cat_1hot = cat_encoder.fit_transform(train_x_b_cat)   #converting the categorical values to numerical ones in order to use in predictions
train_x_b["Street"]=train_x_b_cat_1hot.toarray()    
train_x_d = train_x_b
```

```python
train_x_d
```

### part e

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
train_x_d=scaler.fit_transform(train_x_d)     #scaling values of train dataframe  
train_x_e = pd.DataFrame(data=train_x_d, columns=train_x_b.columns)    #creating new dataframe from scaled values array

```

```python
train_x_e.head(2)
```

## LINEAR REGRESSION TO PREDICT HOUSE PRICES

### part f

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()       
lr_model.fit(train_x_e,train_y)     #performing linear regression for predicting numeric response

from sklearn.metrics import mean_squared_error

SalePrice_predictions = lr_model.predict(train_x_e)
lin_mse = mean_squared_error(train_y, SalePrice_predictions)     #performing mean squared error to find prediction errors
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

### part g

```python
from sklearn.model_selection import cross_val_score
train_mse_score =cross_val_score(lr_model, train_x_e, train_y,     #performing cross validation with 5 segments, evaluating the performance to find mse
                             scoring="neg_mean_squared_error", cv=5)
print(train_mse_score)
```

```python
average_mse = train_mse_score.mean()     #average mse for 5 fold cross validation
print(average_mse)
```

## CLASSIFICATION MODEL TO PREDICT HOUSE PRICE CATEGORY

### part j

```python
train_y_j=train_y
for i in train_y["SalePrice"]:      #labeling the SalePrice train data specifying prices
    if i<100000:
        train_y_j=train_y_j.replace(i,1)
    elif i>=100000 and i<200000:
            train_y_j=train_y_j.replace(i,2)
    elif i>=200000 and i<300000:
        train_y_j=train_y_j.replace(i,3)
    elif i>=300000 and i<400000:
        train_y_j=train_y_j.replace(i,4)
    else:
        train_y_j=train_y_j.replace(i,5)
train_y_j.head(100)
```

```python
test_y_j=test_y
for i in test_y["SalePrice"]:     #labeling test SalePrice data specifying prices
    if i<100000:
        test_y_j=test_y_j.replace(i,1)
    elif i>=100000 and i<200000:
            test_y_j=test_y_j.replace(i,2)
    elif i>=200000 and i<300000:
        test_y_j=test_y_j.replace(i,3)
    elif i>=300000 and i<400000:
        test_y_j=test_y_j.replace(i,4)
    else:
        test_y_j=test_y_j.replace(i,5)
test_y_j.head(100)
```

### part k

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)      #performing random forest classifier for train datas
model.fit(train_x_e, train_y_j)
```

### part l

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
y_train_pred = cross_val_predict(model, train_x_e, train_y_j["SalePrice"], cv=5)    #performing 5 fold cross validation with train data using random forest classifer
```

```python
accuracy =accuracy_score(train_y_j, y_train_pred)     #computing accuracy score, precision, recall, f1 score and confusion matrix for predicted values
precision = precision_score(train_y_j, y_train_pred, average="micro")
recall = recall_score(train_y_j, y_train_pred, average="micro")
f1 = f1_score(train_y_j, y_train_pred, average="micro")
confusion_matrix =confusion_matrix(train_y_j["SalePrice"], y_train_pred)
```

```python
print(confusion_matrix)
print("accuracy: {}\nprecision: {}\nrecall: {}\nf1: {}".format(accuracy,precision,recall,f1)) 
```
